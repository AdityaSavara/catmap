```python
import IPython.html.widgets
import IPython.display
import numpy as np

T, p1, p2 = 600, np.log(1), -1.1


def on_change_T(event, value):
    if type(value) is float:
        global T
        T = value
        #replot(T, p1, p2)


def on_change_p1(event, value):
    if type(value) is float:
        global p1
        p1 = value
        #replot(T, p1, p2)

def on_change_p2(event, value):
    if type(value) is float:
        global p2
        p2 = value
        #replot(T, p1, p2)

def dictify_log(log):
    exec(log)
    return dict(**locals())

def replot(_):
    #print('Replotting ...')
    model.temperature = T
    #model.gas_pressures[1] = np.exp(p1)
    #model.gas_pressures[2] = np.exp(p2)
    #print(model.gas_pressures)
    data, log = model.run()
    info = dictify_log(log)
    #print(sorted(info.keys()))

    xlim, ylim = info['descriptor_ranges']

    x = np.linspace(xlim[0], xlim[1], info['resolution'][0] + 1)
    y = np.linspace(ylim[0], ylim[1], info['resolution'][1] + 1)
    X, Y = np.meshgrid(x, y)
    
    rx, ry = np.array(info['resolution'])
    
    rate = np.zeros((rx + 1, ry + 1))
    CO_coverage = np.zeros((rx + 1, ry + 1))
    O_coverage = np.zeros((rx + 1, ry + 1))

    for (i, j), production_rates in data['production_rate_map']:
        i, j = int((rx) * (i - xlim[0]) / ((xlim[1] - xlim[0]))), int((ry) * (j - ylim[0]) / ((ylim[1] - ylim[0])))
        rate[i, j] = production_rates[0]
    for (i, j), coverages in data['coverage_map']:
        i, j = np.floor((rx) * (i - xlim[0]) / ((xlim[1] - xlim[0]))), np.floor((ry) * (j - ylim[0]) / ((ylim[1] - ylim[0])))
        CO_coverage[i, j] = coverages[0]
        O_coverage[i, j] = coverages[1]

    IPython.display.clear_output(wait=True)
      
    fig = plt.figure(figsize=(16, 9))
    sub = plt.subplot(222); plt.xlim(xlim); plt.ylim(ylim)
        
    sub.title.set_text('CO2 production')
    plt.colorbar(sub.contourf(X, Y, np.log(rate), 20))
    labels = [sub.text(x, y, el) for el, (x, y) in info['descriptor_dict'].items()]
    

    sub = plt.subplot(223); plt.xlim(xlim); plt.ylim(ylim)
    sub.title.set_text('CO coverage')
    labels = [sub.text(x, y, el) for el, (x, y) in info['descriptor_dict'].items()]
    plt.colorbar(sub.contourf(X, Y, CO_coverage, 10))

    sub = plt.subplot(224); plt.xlim(xlim); plt.ylim(ylim)
    sub.title.set_text('O coverage')
    labels = [sub.text(x, y, el) for el, (x, y) in info['descriptor_dict'].items()]
    plt.colorbar(sub.contourf(X, Y, O_coverage, 10))

    
# setup layout
container = IPython.html.widgets.ContainerWidget()
r1_container = IPython.html.widgets.ContainerWidget()
r2_container = IPython.html.widgets.ContainerWidget()
IPython.display.display(container)
T_slider = IPython.html.widgets.FloatSliderWidget(min=300, max=1000, value=T, description='Temperature [K]');# IPython.display.display(T_slider)
T_slider.on_trait_change(on_change_T)


p1_slider = IPython.html.widgets.FloatSliderWidget(min=-9, max=9, value=p1, description='O pressure [bar]');# IPython.display.display(p1_slider)
p1_slider.on_trait_change(on_change_p1)

p2_slider = IPython.html.widgets.FloatSliderWidget(min=-9, max=9, value=p2, description='CO pressure [bar]');# IPython.display.display(p2_slider)
p2_slider.on_trait_change(on_change_p2)

update_button = IPython.html.widgets.ButtonWidget(description='Update!'); #IPython.display.display(update_button)
update_button.on_click(replot)

r1_container.children += (p1_slider, p2_slider,T_slider,);
r2_container.children += (update_button,); 
container.children += (r1_container, r2_container)
container.remove_class('vbox'); container.add_class('hbox')
```
